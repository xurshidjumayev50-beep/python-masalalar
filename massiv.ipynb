```python
# massiv 18 
import numpy as np

n = int(input("n = "))

a = []

for i in range(n):
    son = int(input(f"A[{i}] = "))
    a.append(son)

a = np.array(a)

print("\nMassiv:", a)

oxirgi = a[-1]
print("Oxirgi element:", oxirgi)

kichik = a[a < oxirgi]

if kichik.size > 0:
    print("\nOxirgi elementdan kichik bo‘lgan elementlar:", kichik)
else:
    print("0")
```

```python
# massiv 19 
import numpy as np

n = int(input("n = "))

a = []

for i in range(n):
    son = int(input(f"A[{i}] = "))
    a.append(son)


a = np.array(a)

print("\nMassiv:", a)

oxirgi = a[-1]
print("Oxirgi element:", oxirgi)


kichik = a[a < oxirgi]

# Shunday elementlar bo‘lmasa
if kichik.size == 0:
    print("Oxirgi elementdan kichik bo‘lgan element mavjud emas.")
else:
    eng_katta = np.max(kichik)
    indeks = np.where(a == eng_katta)[0][0]
    print(f"Oxirgi elementdan kichik bo‘lgan eng katta element: {eng_katta}")
    print(f"Indeksi: {indeks}")

```

```python
# massiv 23 
import numpy as np 
n=int(input("n="))
k=int(input("n="))
l=int(input("l="))
a=[]
for _ in range(n):
    a.append(int(input("a = ")))
    
s = 0

for i in range(len(a)):
    if k <= i <= l:
        continue
    else:
        s += a[i]
        

natija=(s/(len(a) - (l-k+1)))
print(natija)
```
